```r
sequences=read.csv("~/Documents/SCTLD/SCTLD_metagenomes/Samples_sequenced.csv")
```

```r
#just keep last 4 digits (year)
sample$Year=sub('.*(?=.{4}$)', '', sample$Month_year, perl=T)
```

```r
sample$SeqID=paste0("T", sample$TransectNum, "_", sample$SampleNum, "_", sample$Year, "_", sample$Species)
```

```r
sample=sample[sample$SeqID %in% sequences$Filename,]
```

```r
sample_colony=merge(sample, colony, by=c("TransectNum", "NewTagNum"))
```

```r
sample_colony$RS=ifelse(sample_colony$Date_DocumentedDisease=="Healthy", "Resistant", "Susceptible")
```

```r
sample_colony
```

```r
resistant=sample_colony[sample_colony$RS=="Resistant",]
susceptible=sample_colony[sample_colony$RS=="Susceptible",]
```
